```python
result = []
```

```python
for index,row in tweets.iterrows():
    try:
        a = userData.loc[(row["user_id"])]
        stance = a["u_stance"]
        if stance=="others":
            stance="other"
        result.append({'u_id': row["user_id"], 'u_sentiment': a["u_sentiment"], 'u_stance': stance})
    except:
        print(".")
```

```python
dataFrame = pd.DataFrame(result)
dataFrame.drop_duplicates(inplace=True)
dataFrame.to_csv('C:/Users/Loris/Documents/Uni/USDA/project/Data/userStance.csv', index=False)
```
